## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item Name"].nunique()
average_price = purchase_data["Price"].mean()
number_purch = purchase_data["Price"].count()
total_rev = purchase_data["Price"].sum()
purch_analysis_tot = pd.DataFrame({"No. of Unique Items":[unique_items], "Ave. Price":[average_price], "No. of Purchases":[number_purch], "Total Revenue":[total_rev]})
purch_analysis_tot["Ave. Price"] = purch_analysis_tot["Ave. Price"].map("${:.2f}".format)
purch_analysis_tot["Total Revenue"] = purch_analysis_tot["Total Revenue"].map("${:,.2f}".format)
purch_analysis_tot = purch_analysis_tot.style.set_properties(**{"text-align":"left"})
purch_analysis_tot
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
user_df = purchase_data.drop_duplicates(subset=["SN"])
user_df
```

```python
total_genders = user_df["Gender"].value_counts()
total_perc = total_genders / total_players
gender_demo = pd.DataFrame({"Total Count":total_genders, "Percentage of Players":total_perc})
gender_demo["Percentage of Players"] = gender_demo["Percentage of Players"].map("{:.2%}".format)
gender_demo = gender_demo.style.set_properties(**{"text-align":"left"})
gender_demo
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
group_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

user_df["Age Ranges"] = pd.cut(user_df["Age"], bins, labels=group_name)
```

```python
age_group = user_df.groupby("Age Ranges")
age_demo_count = user_df["Age Ranges"].value_counts()
age_demo_perc = (age_demo_count / total_players)

age_group = pd.DataFrame({"Total Count": age_demo_count, "Percentage of Players": age_demo_perc})

age_group["Percentage of Players"] = age_group["Percentage of Players"].map("{:.2%}".format)

age_group.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
group_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_name)
purchase_data.head()
```

```python
purch_analysis = purchase_data.groupby("Age Ranges")


purch_count = purchase_data.groupby(["Age Ranges"]).count()["Price"]
avg_purch_price = purchase_data.groupby(["Age Ranges"]).mean()["Price"]
total_purch = purchase_data.groupby(["Age Ranges"]).sum()["Price"]
avg_total_purch = total_purch / purch_count


purch_analysis = pd.DataFrame({"Purchase Count": purch_count,
                               "Average Purchase Price": avg_purch_price, 
                               "Total Purchase Value": total_purch,
                               "Avg Total Purchase per Person": avg_total_purch})


purch_analysis["Average Purchase Price"] = purch_analysis["Average Purchase Price"].map("${:.2f}".format)
purch_analysis["Total Purchase Value"] = purch_analysis["Total Purchase Value"].map("${:.2f}".format)
purch_analysis["Avg Total Purchase per Person"] = purch_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

purch_analysis.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = purchase_data.groupby("SN")

purch_count = top_spender["Purchase ID"].count()
avg_purch_top_spend = top_spender["Price"].mean()
purch_total_top_spend = top_spender["Price"].sum()

top_spend_df = pd.DataFrame({"Purchase Count": purch_count,
                            "Average Purchase Price": avg_purch_top_spend,
                            "Total Purchase Value": purch_total_top_spend})

format_top_spend = top_spend_df.sort_values(["Total Purchase Value"], ascending=False).head()

format_top_spend.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items = purchase_data[["Item ID", "Item Name", "Price"]]

item_stats = pop_items.groupby(["Item ID", "Item Name"])

item_count = item_stats["Price"].count()
tot_purch_value = item_stats["Price"].sum()
item_price = tot_purch_value / item_count

pop_itmes_df = pd.DataFrame({"Purchase Count": item_count,
                            "Item Price": item_price,
                            "Total Purchase Value": tot_purch_value})

pop_items_df_format = pop_itmes_df.sort_values(["Purchase Count"], ascending=False).head()
pop_items_df_format.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})

```
